```python
# ============================================
# PlantDocX — Single Image Disease Prediction
# ============================================
import os, json, pickle, warnings, datetime
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

# --------- Paths ----------
IMAGE_PATH = r"C:\Users\sagni\Downloads\PlantDocX\download.jpg"
OUTPUT_DIR = r"C:\Users\sagni\Downloads\PlantDocX"
PP_PATH    = os.path.join(OUTPUT_DIR, "preprocessor.pkl")
KERAS_PATH = os.path.join(OUTPUT_DIR, "cls_model.keras")
H5_PATH    = os.path.join(OUTPUT_DIR, "cls_model.h5")
os.makedirs(OUTPUT_DIR, exist_ok=True)

# --------- Load preprocessor (class names, image size, etc.) ----------
if not os.path.exists(PP_PATH):
    raise FileNotFoundError(f"Missing preprocessor: {PP_PATH}")

with open(PP_PATH, "rb") as f:
    preproc = pickle.load(f)

class_names = preproc.get("class_names")
if not class_names:
    raise RuntimeError("class_names missing in preprocessor.pkl")

IMG_SIZE = tuple(preproc.get("image_size", (224, 224)))
RESCALE_IN_MODEL = bool(preproc.get("rescale_in_model", True))  # set True if model had Rescaling(1/255)

# --------- Load model (.keras preferred, fallback to .h5) ----------
model = None
if os.path.exists(KERAS_PATH):
    try:
        model = tf.keras.models.load_model(KERAS_PATH, safe_mode=False)
        print(f"[INFO] Loaded model: {KERAS_PATH}")
    except Exception as e:
        print("[WARN] .keras load failed:", e)

if model is None and os.path.exists(H5_PATH):
    try:
        model = tf.keras.models.load_model(H5_PATH, compile=False)
        print(f"[INFO] Loaded legacy model: {H5_PATH}")
    except Exception as e:
        raise RuntimeError(f"Could not load any model: {e}")

# --------- Load & preprocess image ----------
if not os.path.exists(IMAGE_PATH):
    raise FileNotFoundError(f"Image not found: {IMAGE_PATH}")

img = tf.keras.utils.load_img(IMAGE_PATH, target_size=IMG_SIZE)
x = tf.keras.utils.img_to_array(img)
if not RESCALE_IN_MODEL:
    x = x / 255.0
x = np.expand_dims(x, axis=0)

# --------- Predict ----------
probs = model.predict(x, verbose=0)[0]  # shape: (num_classes,)
num_classes_model = probs.shape[-1]
if len(class_names) != num_classes_model:
    # Safety fallback: align class names to model output if mismatch
    print(f"[WARN] class_names len ({len(class_names)}) != model outputs ({num_classes_model}); generating fallback names.")
    class_names = [f"class_{i}" for i in range(num_classes_model)]

pred_idx = int(np.argmax(probs))
pred_name = class_names[pred_idx]
pred_conf = float(probs[pred_idx])

# Identify "Healthy" vs "Diseased" (based on class name containing 'healthy')
healthy_indices = [i for i, n in enumerate(class_names) if "healthy" in n.lower()]
healthy_set = set(healthy_indices)
is_healthy = pred_idx in healthy_set if healthy_indices else False
binary_label = "Healthy" if is_healthy else "Diseased"

# Top-5
top_k = min(5, len(class_names))
top_idx = np.argsort(probs)[::-1][:top_k]
top_pairs = [(class_names[i], float(probs[i])) for i in top_idx]

print("\n=== PlantDocX Prediction ===")
print(f"Image: {IMAGE_PATH}")
print(f"Prediction: {pred_name}  (p={pred_conf:.4f})  -> {binary_label}")
print("Top-5:")
for name, p in top_pairs:
    print(f" - {name}: {p:.4f}")

# --------- Show results in Jupyter (image + top-5 bar chart) ----------
# Figure 1: Image with title
plt.figure(figsize=(6,6))
plt.imshow(tf.keras.utils.array_to_img(x[0] if RESCALE_IN_MODEL else (x[0]*255.0).astype(np.uint8)))
plt.axis('off')
plt.title(f"{pred_name} ({pred_conf:.2%}) — {binary_label}")
plt.show()

# Figure 2: Top-5 bar chart
labels = [class_names[i] for i in top_idx]
values = [probs[i] for i in top_idx]
plt.figure(figsize=(7,4))
plt.barh(range(top_k), values[::-1])
plt.yticks(range(top_k), labels[::-1])
plt.xlabel("Probability")
plt.title("Top-5 probabilities")
plt.tight_layout()
plt.show()

# --------- Save artifacts ----------
base = os.path.splitext(os.path.basename(IMAGE_PATH))[0]
pred_img_path = os.path.join(OUTPUT_DIR, f"{base}_prediction.png")
top5_path     = os.path.join(OUTPUT_DIR, f"{base}_top5.png")
json_path     = os.path.join(OUTPUT_DIR, f"{base}_prediction.json")

# Save the display images again (without re-showing)
# Image with title
plt.figure(figsize=(6,6))
plt.imshow(tf.keras.utils.array_to_img(x[0] if RESCALE_IN_MODEL else (x[0]*255.0).astype(np.uint8)))
plt.axis('off')
plt.title(f"{pred_name} ({pred_conf:.2%}) — {binary_label}")
plt.tight_layout()
plt.savefig(pred_img_path, dpi=150)
plt.close()

# Top-5 chart
plt.figure(figsize=(7,4))
plt.barh(range(top_k), values[::-1])
plt.yticks(range(top_k), labels[::-1])
plt.xlabel("Probability")
plt.title("Top-5 probabilities")
plt.tight_layout()
plt.savefig(top5_path, dpi=150)
plt.close()

# JSON
result = {
    "image_path": IMAGE_PATH,
    "pred_class": pred_name,
    "pred_probability": round(pred_conf, 6),
    "binary_decision": binary_label,
    "top_k": [{"class": cn, "prob": round(float(p), 6)} for cn, p in top_pairs],
    "timestamp": datetime.datetime.now().isoformat(timespec="seconds")
}
with open(json_path, "w", encoding="utf-8") as f:
    json.dump(result, f, indent=2)

print(f"\n[INFO] Saved:")
print(f" - {pred_img_path}")
print(f" - {top5_path}")
print(f" - {json_path}")
```
