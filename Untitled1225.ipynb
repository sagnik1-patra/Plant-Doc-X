```python
# =========================================================
# PlantDocX  Potato Disease Prediction + Robust CAM (Grad-CAM -> Score-CAM fallback)
# =========================================================
import os, json, pickle, warnings, datetime
warnings.filterwarnings("ignore")

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import tensorflow as tf
from tensorflow import keras

# ---------- Paths ----------
IMAGE_PATH = r"C:\Users\sagni\Downloads\PlantDocX\download.jpg"   # your potato photo
OUTPUT_DIR = r"C:\Users\sagni\Downloads\PlantDocX"
PP_PATH    = os.path.join(OUTPUT_DIR, "preprocessor.pkl")
KERAS_PATH = os.path.join(OUTPUT_DIR, "cls_model.keras")
H5_PATH    = os.path.join(OUTPUT_DIR, "cls_model.h5")
os.makedirs(OUTPUT_DIR, exist_ok=True)

# ---------- Load preprocessor ----------
with open(PP_PATH, "rb") as f:
    preproc = pickle.load(f)

class_names = list(preproc.get("class_names", []))
if not class_names:
    raise RuntimeError("class_names missing in preprocessor.pkl")

IMG_SIZE = tuple(preproc.get("image_size", (224, 224)))
RESCALE_IN_MODEL = bool(preproc.get("rescale_in_model", True))

# ---------- Load model ----------
model = None
if os.path.exists(KERAS_PATH):
    try:
        model = tf.keras.models.load_model(KERAS_PATH, safe_mode=False)
        print(f"[INFO] Loaded model: {KERAS_PATH}")
    except Exception as e:
        print("[WARN] .keras load failed:", e)

if model is None and os.path.exists(H5_PATH):
    model = tf.keras.models.load_model(H5_PATH, compile=False)
    print(f"[INFO] Loaded legacy model: {H5_PATH}")

# ---------- Load & preprocess image ----------
if not os.path.exists(IMAGE_PATH):
    raise FileNotFoundError(f"Image not found: {IMAGE_PATH}")

pil = Image.open(IMAGE_PATH).convert("RGB")
pil = ImageOps.exif_transpose(pil)
min_side = min(pil.size)
left = (pil.width - min_side)//2
top  = (pil.height - min_side)//2
pil_square = pil.crop((left, top, left+min_side, top+min_side))
pil_resized = pil_square.resize(IMG_SIZE, Image.BICUBIC)

x_uint8 = tf.keras.utils.img_to_array(pil_resized).astype(np.uint8)
x = x_uint8.astype("float32")
x_in = np.expand_dims(x, 0)
if not RESCALE_IN_MODEL:
    x_in = x_in / 255.0

# ---------- Predict ----------
probs = model.predict(x_in, verbose=0)[0]
C = probs.shape[-1]
if len(class_names) != C:
    print(f"[WARN] class_names len ({len(class_names)}) != model outputs ({C}); using fallback names.")
    class_names = [f"class_{i}" for i in range(C)]

# Prefer Potato if present
potato_idxs = [i for i, n in enumerate(class_names) if "potato" in n.lower()]
global_idx  = int(np.argmax(probs))
global_name = class_names[global_idx]
global_prob = float(probs[global_idx])

if potato_idxs:
    pick_idx = int(potato_idxs[np.argmax(probs[potato_idxs])])
    pick_name = class_names[pick_idx]
    pick_prob = float(probs[pick_idx])
    crop_hint = "Potato (forced preference when available)"
else:
    pick_idx, pick_name, pick_prob = global_idx, global_name, global_prob
    crop_hint = "No explicit Potato class in outputs; using global top-1"

healthy_idxs = [i for i, n in enumerate(class_names) if "healthy" in n.lower()]
is_healthy = (pick_idx in healthy_idxs) if healthy_idxs else ("healthy" in pick_name.lower())
binary_label = "Healthy" if is_healthy else "Diseased"

top_k = min(5, C)
top_idx = np.argsort(probs)[::-1][:top_k]
top_pairs = [(class_names[i], float(probs[i])) for i in top_idx]

print("\n=== PlantDocX Potato Prediction ===")
print(f"Image: {IMAGE_PATH}")
print(f"Chosen: {pick_name}  (p={pick_prob:.4f})  -> {binary_label}")
print(f"Note: {crop_hint}")
print("Top-5:")
for name, p in top_pairs:
    print(f" - {name}: {p:.4f}")

# ---------- Layer helpers ----------
ConvLike = (tf.keras.layers.Conv2D,
            tf.keras.layers.SeparableConv2D,
            tf.keras.layers.DepthwiseConv2D)

def iter_all_layers(m):
    for l in m.layers:
        yield l
        if isinstance(l, tf.keras.Model):
            for sl in iter_all_layers(l):
                yield sl

def get_last_conv_layer(model):
    last = None
    for l in iter_all_layers(model):
        try:
            rank = len(l.output.shape)
        except Exception:
            continue
        if rank == 4 and isinstance(l, ConvLike):
            last = l
    return last

last_conv = get_last_conv_layer(model)
heatmap = None
cam_method = None

# ---------- Try Grad-CAM ----------
if last_conv is not None:
    try:
        grad_model = tf.keras.models.Model(
            [model.inputs], [last_conv.output, model.output]
        )
        with tf.GradientTape() as tape:
            conv_out, preds = grad_model(x_in)
            class_channel = preds[:, pick_idx]
        grads = tape.gradient(class_channel, conv_out)

        if grads is not None:
            pooled = tf.reduce_mean(grads, axis=(1, 2), keepdims=True)
            cam = tf.nn.relu(tf.reduce_sum(pooled * conv_out, axis=-1))[0].numpy()
            cam -= cam.min() if cam.max() > cam.min() else 0.0
            cam /= (cam.max() + 1e-8)
            heatmap = tf.image.resize(cam[..., None], IMG_SIZE).numpy()[..., 0]
            cam_method = "Grad-CAM"
        else:
            print("[WARN] Gradients were None. Will try Score-CAM fallback.")
    except Exception as e:
        print("[WARN] Grad-CAM failed:", e)

# ---------- Score-CAM fallback (no gradients) ----------
def score_cam(x_in, conv_layer, top_channels=64):
    # Forward once to get conv maps and their importance
    probe = tf.keras.models.Model([model.inputs], [conv_layer.output, model.output])
    conv_maps, _ = probe(x_in)
    conv_maps = conv_maps[0].numpy()        # (H, W, K)
    H, W, K = conv_maps.shape

    # pick top channels by global activation
    act = conv_maps.reshape(-1, K).mean(axis=0)
    idxs = np.argsort(act)[::-1][:min(top_channels, K)]

    # build weights by masking input with each upsampled map
    weights = []
    for k in idxs:
        m = conv_maps[..., k]
        m = (m - m.min()) / (m.max() - m.min() + 1e-8)
        m_up = tf.image.resize(m[..., None], IMG_SIZE).numpy()[..., 0]  # (Himg, Wimg)
        # Mask the input (broadcast to 3 channels)
        masked = (x / 255.0 if not RESCALE_IN_MODEL else x).copy()
        masked *= m_up[..., None].astype("float32")
        masked = np.expand_dims(masked, 0)
        p = model.predict(masked, verbose=0)[0, pick_idx]
        weights.append(float(p))

    weights = np.array(weights, dtype="float32")
    weights = np.maximum(weights, 0)
    if weights.sum() == 0:
        weights = np.ones_like(weights)

    # combine selected maps with weights
    cam = np.zeros((H, W), dtype="float32")
    for w, k in zip(weights, idxs):
        cam += w * conv_maps[..., k]
    cam = np.maximum(cam, 0)
    cam -= cam.min() if cam.max() > cam.min() else 0.0
    cam /= (cam.max() + 1e-8)
    cam_up = tf.image.resize(cam[..., None], IMG_SIZE).numpy()[..., 0]
    return cam_up

if heatmap is None and last_conv is not None:
    try:
        heatmap = score_cam(x_in, last_conv, top_channels=64)
        cam_method = "Score-CAM"
        print("[INFO] Used Score-CAM fallback.")
    except Exception as e:
        print("[WARN] Score-CAM failed:", e)

# ---------- Show inline ----------
plt.figure(figsize=(6,6))
plt.imshow(pil_resized)
plt.axis("off")
plt.title(f"{pick_name} ({pick_prob:.2%})  {binary_label}")
plt.show()

plt.figure(figsize=(7,4))
labels = [class_names[i] for i in top_idx][::-1]
vals   = [probs[i] for i in top_idx][::-1]
plt.barh(range(top_k), vals)
plt.yticks(range(top_k), labels)
plt.xlabel("Probability")
plt.title("Top-5 probabilities")
plt.tight_layout()
plt.show()

if heatmap is not None:
    # overlay without OpenCV (pure matplotlib/PIL)
    cmap = plt.get_cmap("jet")
    hm_rgb = (cmap(heatmap)[..., :3] * 255).astype(np.uint8)
    overlay = (0.6 * x_uint8 + 0.4 * hm_rgb).astype(np.uint8)

    plt.figure(figsize=(6,6))
    plt.imshow(overlay)
    plt.axis("off")
    plt.title(f"{cam_method} overlay: {pick_name}")
    plt.tight_layout()
    plt.show()
else:
    print("[WARN] No heatmap generated (no suitable conv layer or CAM failed).")

# ---------- Save outputs ----------
base = os.path.splitext(os.path.basename(IMAGE_PATH))[0]
pred_png  = os.path.join(OUTPUT_DIR, f"{base}_potato_prediction.png")
top5_png  = os.path.join(OUTPUT_DIR, f"{base}_top5.png")
json_path = os.path.join(OUTPUT_DIR, f"{base}_prediction.json")
grad_png  = os.path.join(OUTPUT_DIR, f"{base}_cam.png") if heatmap is not None else None

# Save prediction image
plt.figure(figsize=(6,6))
plt.imshow(pil_resized)
plt.axis("off")
plt.title(f"{pick_name} ({pick_prob:.2%})  {binary_label}")
plt.tight_layout()
plt.savefig(pred_png, dpi=150); plt.close()

# Save top-5
plt.figure(figsize=(7,4))
plt.barh(range(top_k), vals)
plt.yticks(range(top_k), labels)
plt.xlabel("Probability")
plt.title("Top-5 probabilities")
plt.tight_layout()
plt.savefig(top5_png, dpi=150); plt.close()

# Save CAM overlay
if heatmap is not None:
    cmap = plt.get_cmap("jet")
    hm_rgb = (cmap(heatmap)[..., :3] * 255).astype(np.uint8)
    overlay = (0.6 * x_uint8 + 0.4 * hm_rgb).astype(np.uint8)
    Image.fromarray(overlay).save(grad_png)

# Save JSON
result = {
    "image_path": IMAGE_PATH,
    "chosen_class": pick_name,
    "chosen_probability": round(pick_prob, 6),
    "binary_decision": binary_label,
    "note": crop_hint,
    "global_top1": {"class": global_name, "prob": round(global_prob, 6)},
    "top_k": [{"class": cn, "prob": round(float(p), 6)} for cn, p in top_pairs],
    "cam_method": cam_method,
    "timestamp": datetime.datetime.now().isoformat(timespec="seconds"),
    "artifacts": {
        "prediction_png": pred_png,
        "top5_png": top5_png,
        "cam_png": grad_png
    }
}
with open(json_path, "w", encoding="utf-8") as f:
    json.dump(result, f, indent=2)

print("\n[INFO] Saved:")
print(" -", pred_png)
print(" -", top5_png)
if grad_png: print(" -", grad_png, f"({cam_method})")
print(" -", json_path)
```
